```python
asianEngines = df.loc[df['origin'] == 'Asia']
df_asian = asianEngines.head(7)
df_asian = df_asian[['vehicle_weight', 'model_year']]
X = df_asian.to_numpy()
print(X)
```

```python
XTX = X.T @ X
print(XTX)
```

```python
XTX_inv = np.linalg.inv(XTX)
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
w = (XTX_inv @ X.T) @ y
print(np.sum(w))
```
